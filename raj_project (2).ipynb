```python
age_group_1 = '10-19'
age_group_2 = '20-29'
filtered_df = df[df['Age-group'].isin([age_group_1, age_group_2])] 
sum_age_group_1 = filtered_df[filtered_df['Age-group'] == age_group_1]['In seeing'].sum()
sum_age_group_2 = filtered_df[filtered_df['Age-group'] == age_group_2]['In seeing'].sum()

# the total sum 
total_sum =df['In seeing'].sum() 
# Calculate percentages relative to the total sum
percentage_age_group_1 = (sum_age_group_1 / total_sum) * 100
percentage_age_group_2 = (sum_age_group_2 / total_sum) * 100
print(f"Sum and Percentage of In Seeing for Specific Age Groups:")
print(f"  Age Group: {age_group_1} - Sum: {sum_age_group_1:.2f}, Percentage: {percentage_age_group_1:.2f}%")
print(f"  Age Group: {age_group_2} - Sum: {sum_age_group_2:.2f}, Percentage: {percentage_age_group_2:.2f}%")
```

```python
data_to_plot = [percentage_age_group_1, percentage_age_group_2]
age_group_labels = [age_group_1, age_group_2]
plt.figure(figsize=(6, 4)) 
sns.barplot(x=age_group_labels, y=data_to_plot)
plt.xlabel("Age Group")
plt.ylabel("Percentage (%)")
plt.title("Percentage of In Seeing for Specific Age Groups")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
